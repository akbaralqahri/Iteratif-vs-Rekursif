```python
import matplotlib.pyplot as plt
```

MENAMPILKAN KOMPLEKSITAS WAKTU PROGRAM

```python
def sumofpoweriterative2(n):
    return 2 * n + 2

def sumofpowerrecursive2(n):
    return n
```

MENAMPILKAN KOMPLEKSITAS WAKTU ALGORITMA REKURSIF

```python
rekursif = []
n = 100

for _ in range(1, 5):
    rekursif.append(sumofpowerrecursive2(n))
    n = n + n

#Grafik
plt.title("Kompleksitas Waktu Algoritma Rekursif", fontsize=18,)

plt.plot(rekursif, color="red", linewidth=2.0, linestyle="-", label = 'rekursif')
# plt.plot(x3, y3, color="red", linewidth=2.0, linestyle="-", label = 'bagus')
plt.legend(frameon = False)
plt.show()
print(rekursif)
```

MENAMPILKAN KOMPLEKSITAS WAKTU ALGORITMA ITERATIF

```python
iteratif = []
n = 100

for _ in range(1, 5):
    iteratif.append(sumofpoweriterative2(n))
    n = n + n

#Grafik
plt.title("Kompleksitas Waktu Algoritma Iteratif", fontsize=18,)

plt.plot(iteratif, color="black", linewidth=2.0, linestyle="-", label = 'iteratif')

plt.legend(frameon = False)
plt.show()
print(iteratif)
```

MENAMPILKAN PERBANDINGAN KOMPLEKSITAS WAKTU REKURSIF DAN ITERATIF

```python
rekursif = [0]
iteratif = [0]
n = 100
m = 100

for i in range(1, 5):
    rekursif.append(sumofpowerrecursive2(n))
    n = n + n

for i in range(1, 5):
    iteratif.append(sumofpoweriterative2(m))
    m = m + m

print('berikut isi dari list rekursif\n', rekursif)
print('berikut isi dari list iteratif\n', iteratif)

#Grafik
plt.title("Perbandingan Kompleksitas Waktu", fontsize=15,)

plt.plot(rekursif, color="red", linewidth=2.5, linestyle="-", label = 'rekursif')
plt.plot(iteratif, color="black", linewidth=2.0, linestyle="-", label = 'iteratif')
# plt.plot(x3, y3, color="red", linewidth=2.0, linestyle="-", label = 'bagus')
plt.legend(frameon = False)
plt.show()
```

MENAMPILKAN OUTPUT DARI PROGRAM

```python
def sumofpoweriterative(n):
    sum = 0
    i = 0
    for i in range(n):
        sum = sum + i * i
    return sum

def sumofpowerrecursive(n):
    if n == 1:
        return 1
    else:
        return n * n + sumofpowerrecursive(n-1)
```

MENAMPILKAN GRAFIK OUTPUT ALGORITMA REKURSIF

```python
rekursif = []
n = 100

for _ in range(1, 5):
    rekursif.append(sumofpowerrecursive(n))
    n = n + n

#Grafik
plt.title("Output Algoritma Rekursif", fontsize=18,)

plt.plot(rekursif, color="red", linewidth=2.0, linestyle="-", label = 'rekursif')
# plt.plot(x3, y3, color="red", linewidth=2.0, linestyle="-", label = 'bagus')
plt.legend(frameon = False)
plt.show()
print(rekursif)
```

MENAMPILKAN GRAFIK OUTPUT ALGORITMA ITERATIF

```python
iteratif = []
n = 100

for _ in range(1, 5):
    iteratif.append(sumofpoweriterative(n))
    n = n + n

#Grafik
plt.title("Output Algoritma Iteratif", fontsize=18,)

plt.plot(iteratif, color="black", linewidth=2.0, linestyle="-", label = 'iteratif')

plt.legend(frameon = False)
plt.show()
print(iteratif)
```

MENAMPILKAN PERBANDINGAN OUTPUT ANTARA ALGORITMA REKURSIF DAN ALGORITMA ITERATIF

```python
rekursif = [0]
iteratif = [0]
n = 100
m = 100

for i in range(1, 5):
    rekursif.append(sumofpowerrecursive(n))
    n = n + n

for i in range(1, 5):
    iteratif.append(sumofpoweriterative(m))
    m = m + m

print('berikut isi dari list rekursif\n', rekursif)
print('berikut isi dari list iteratif\n', iteratif)

#Grafik
plt.title("Perbandingan Output", fontsize=15,)

plt.plot(rekursif, color="red", linewidth=2.5, linestyle="--", label = 'rekursif')
plt.plot(iteratif, color="black", linewidth=2.0, linestyle="--", label = 'iteratif')
# plt.plot(x3, y3, color="red", linewidth=2.0, linestyle="-", label = 'bagus')
plt.legend(frameon = False)
plt.show()

```
