```python
import matplotlib.pyplot as plt
```

```python
def sumofpoweriterative2(n):
    return 2 * n + 2

def sumofpowerrecursive2(n):
    return n
```

MENAMPILKAN KOMPLEKSITAS WAKTU ALGORITMA REKURSIF

```python
rekursif = []
n = 100

for _ in range(1, 5):
    rekursif.append(sumofpowerrecursive2(n))
    n = n + n

#Grafik
plt.title("Kompleksitas Waktu Algoritma Rekursif", fontsize=18,)

plt.plot(rekursif, color="red", linewidth=2.0, linestyle="-", label = 'rekursif')
# plt.plot(x3, y3, color="red", linewidth=2.0, linestyle="-", label = 'bagus')
plt.legend(frameon = False)
plt.show()
print(rekursif)
```

MENAMPILKAN KOMPLEKSITAS WAKTU ALGORITMA ITERATIF

```python
iteratif = []
n = 100

for _ in range(1, 5):
    iteratif.append(sumofpoweriterative2(n))
    n = n + n

#Grafik
plt.title("Kompleksitas Waktu Algoritma Iteratif", fontsize=18,)

plt.plot(iteratif, color="black", linewidth=2.0, linestyle="-", label = 'iteratif')

plt.legend(frameon = False)
plt.show()
print(iteratif)
```

MENAMPILKAN GRAFIK PERBANDINGAN WAKTU ANTARA ALGORITMA REKURSIF DAN ITERATIF

```python
rekursif = [0]
iteratif = [0]
n = 100
m = 100

for i in range(1, 5):
    rekursif.append(sumofpowerrecursive2(n))
    n = n + n

for i in range(1, 5):
    iteratif.append(sumofpoweriterative2(m))
    m = m + m

print('berikut isi dari list rekursif\n', rekursif)
print('berikut isi dari list iteratif\n', iteratif)

#Grafik
plt.title("Perbandingan Kompleksitas Waktu", fontsize=15,)

plt.plot(rekursif, color="red", linewidth=2.5, linestyle="--", label = 'rekursif')
plt.plot(iteratif, color="black", linewidth=2.0, linestyle="--", label = 'iteratif')
# plt.plot(x3, y3, color="red", linewidth=2.0, linestyle="-", label = 'bagus')
plt.legend(frameon = False)
plt.show()
```
